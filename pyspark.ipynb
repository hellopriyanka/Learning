```python
taxi_jan2025=spark.read.parquet('/content/drive/MyDrive/PySpark/yellow_tripdata_2025-01.parquet')
taxi_jan2025.createOrReplaceTempView('taxi_jan2025')
```

```python
taxi_lookup=spark.read.option('header','true')\
.csv('/content/drive/MyDrive/PySpark/taxi_zone_lookup.csv')
taxi_lookup.createOrReplaceTempView('taxi_lookup')
```

```python
QUERY='''
SELECT DOLocationID, Borough, total_amount
FROM taxi_jan2025
LEFT JOIN taxi_lookup
ON DOLocationID=LocationID
'''
joined_df=spark.sql(query)
```

```python
joined_df.groupBy('Borough') \
    .agg(avg('total_amount').alias('avg_amount')) \
    .show()
```
